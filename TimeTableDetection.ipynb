```python
import fitz  # PyMuPDF

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text_by_page = []
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text = page.get_text("text")
        text_by_page.append(text)
    return text_by_page

# Function to process text and generate busy slots report
def process_timetable(text_by_page):
    report = []
    for page_num, page_text in enumerate(text_by_page):
        lines = page_text.splitlines()
        current_class = None
        times = []
        days = ["Mo", "Tu", "We", "Th", "Fr"]
        day_slots = {day: [] for day in days}

        for line in lines:
            line = line.strip()

            # Identify class number
            if line.startswith("A") and "-" in line:
                current_class = line
                continue

            # Identify time slots
            if "-" in line and ":" in line and not any(day in line for day in days):
                times = line.split()
                continue

            # Process class slots
            if any(day in line for day in days):
                current_day = [day for day in days if day in line][0]
                slots = line.split()
                for idx, slot in enumerate(slots):
                    if slot and idx < len(times):  # Ensure index is within range
                        if slot != '-':  # Skip empty slots
                            day_slots[current_day].append(idx + 1)  # Add 1 to make slot numbers start from 1
                    elif idx >= len(times):
                        print(f"Warning: Slot index {idx} out of range for times {times}")

        if current_class:
            report.append(f"Class {current_class}:")
            for day in days:
                busy_slots = day_slots[day]
                if busy_slots:
                    report.append(f"{day}: {' '.join(map(str, busy_slots))}")
                else:
                    report.append(f"{day}: No busy slots")
            report.append("")  # Add a blank line between classes

    return "\n".join(report)


# Upload the PDF to Colab environment
from google.colab import files

uploaded = files.upload()

# Get the path of the uploaded file
pdf_path = next(iter(uploaded))

# Extract text from the uploaded PDF
text_by_page = extract_text_from_pdf(pdf_path)

# Generate the busy slots report
report = process_timetable(text_by_page)

# Print the report
print(report)
```
